```python
import numpy as np
from numpy import arctan2 as atan2, sin, cos
import matplotlib.pyplot as plt
from IPython import display
from scipy.constants import Boltzmann as kB
from tkinter import *
from tkinter import ttk
from PIL import ImageGrab
import time
import math

# parameters
N = 50  # no. of particles
L = 20  # length
r = 2  # radius
dt = 0.4  # timestep
# eps = 0.001 #repulsion
v_ab = 10  # absolute velocity
c1 = 3  # binocualar
c2 = 2  # onocular
eta = 0.4  # noise
n1 = 0
n2 = 0
Nt = 10000
plotRealTime = True

res = 500  # Resolution of the animation
tk = Tk()
tk.geometry(str(int(res * 1.1)) + 'x' + str(int(res * 1.3)))
tk.configure(background='white')

canvas = Canvas(tk, bd=2)  # Generate animation window
tk.attributes('-topmost', 0)
canvas.place(x=res / 20, y=res / 20, height=res, width=res)
ccolor = ['#17888E', '#C1D02B', '#9E00C9', '#D80000', '#E87B00', '#9F68D3', '#4B934F']



# loc = [x.tolist(),y.tolist()]
# loc = np.array([x,y])

# velocity positions , velocity is suppose to null
theta = np.random.rand(N, 1) * 2 * np.pi
theta1 = np.round_(theta, decimals=0)
# print(theta1)
v = [v_ab * np.cos(theta), v_ab * np.sin(theta)]

# distance matrix
dis_mat = [[] for x in range(N)]
ed = [[] for x in range(N)]


# distance calculation function
def calc(x1, y1, x2, y2):
    po = round(math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2), 2)
    return po


def dismat(x, y):
    # distance matrix
    dis_mat = [[] for x in range(N)]
    ed = [[] for x in range(N)]
    # dismat matrix update
    for j in range(N):
        for i in range(N):
            dis_mat[j].append(calc(x[i], y[i], x[j], y[j]))

    dis_mat = np.array(dis_mat)

    # less than r only
    ed = (dis_mat < r).astype(int)
    ed = np.array(ed)

    return dis_mat, ed


def twopi(x):
    if x < 0:
        return 2 * np.pi + x
    if x >= 0:
        return x


# bird location function
ran = np.zeros((N, N))
ran_copy = np.zeros((N, N))
v_avgx = np.zeros((N))
v_avgy = np.zeros((N))

# Parameters of the simulation
n = 50  # Number of particles
dt = 0.03  # Time step

# Physical parameters of the system
l = 100
V = 20  # Particle velocity
R = 4
Dr = 0.4  # noise

x = np.random.rand(n) * 2 * l - l  # x coordinates
y = np.random.rand(n) * 2 * l - l  # y coordinates
# phi = np.random.rand(n) * 2 * np.pi  # orientations                # Initialization

particles = []
for j in range(N):  # Generate animated particles in Canvas
    particles.append(canvas.create_oval((x[j] - R + l) * res / l / 2, \
                                        (y[j] - R + l) * res / 2 / l, \
                                        (x[j] + R + l) * res / l / 2, \
                                        (y[j] + R + l) * res / l / 2, \
                                        outline=ccolor[0], fill=ccolor[0]))

for run in range(Nt):
    mean_theta = theta
    dis_mat, ed = dismat(x,y)    


    for i in range(N):

        x = (x + v[0][i] * dt + l) % (2 * l) - l  # Update x coordinates
        y = (y + v[1][i] * dt + l) % (2 * l) - l  # Update y coordinates
        bino_velx = 0
        bino_vely = 0  # for velocity sum
        ocu_velx = 0
        ocu_vely = 0
        n1 = 0
        n2 = 0

        for j in range(N):
            if ed[i][j] == 1 and i != j:
                ran[i][j] = (np.arctan((y[j] - y[i]) / (x[j] - x[i])))
                ran[i][j] = round(ran[i][j])
                ran[i][j] = twopi(ran[i][j])  # change it to 2pi degree

                if 0 <= abs(ran[i][j] - theta1[i]) < (11.0 * np.pi / 180) or (349 * np.pi / 180) <= abs(
                        ran[i][j] - theta1[i]) <= (360 * np.pi / 180):
                    # print(abs(ran[i][j] - theta1[i]),ran[i][j],theta1[i])
                    ran_copy[i][j] = 1  # binocular region
                    bino_velx = bino_velx + v[0][j]
                    bino_vely = bino_vely + v[1][j]
                    n1 = n1 + 1

                elif (11 * np.pi / 180) <= abs(ran[i][j] - theta1[i]) < (153 * np.pi / 180) or (
                        207 * np.pi / 180) <= abs(ran[i][j] - theta1[i]) <= (349 * np.pi / 180):
                    #                print(abs(ran[i][j] - theta1[i]),ran[i][j],theta1[i])
                    ran_copy[i][j] = -1  # monocular region
                    ocu_velx = ocu_velx + v[0][j]
                    ocu_vely = ocu_vely + v[1][j]
                #                n2[i] = n2[i] + 1

                elif (153 * np.pi / 180) <= abs(ran[i][j] - theta1[i]) < (207 * np.pi / 180):
                    #                print(abs(ran[i][j] - theta1[i]),ran[i][j],theta1[i])
                    ran_copy[i][j] = 0  # visionless region

        coeff = 1 / (c1 * n1 + c2 * (1 - n1))
        v_avgx = coeff * (c1 * bino_velx + c2 * ocu_velx)
        v_avgy = coeff * (c1 * bino_vely + c2 * ocu_vely)

        mean_theta[i] = np.arctan2(v_avgy, v_avgx)
        theta[i] = mean_theta[i] + eta * (np.random.random() - 0.5)  # updating theta values
        v[0][i] = v_ab * np.cos(theta[i])
        v[1][i] = v_ab * np.sin(theta[i])

    for j in range(N):
        canvas.coords(particles[j],
                      (x[j] - R + l) * res / l / 2, \
                      (y[j] - R + l) * res / 2 / l, \
                      (x[j] + R + l) * res / l / 2, \
                      (y[j] + R + l) * res / l / 2, )  # Updating animation coordinates
    tk.title('t =' + str(round(i * dt * 100) / 100))  # Animation title
    tk.update()  # Update animation frame
    time.sleep(0.01)  # Wait between loops

Tk.mainloop(canvas)  # Release animation handle (close window to finish)
```
